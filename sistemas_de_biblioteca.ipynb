```python
lista_livros = []

class Livro:
    def __init__(self, titulo, autor, genero, quantidade_disponivel):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade_disponivel = quantidade_disponivel

    def __str__(self):
        return f"Título: {self.titulo}, Autor: {self.autor}, Gênero: {self.genero}, Quantidade disponível: {self.quantidade_disponivel}"

livro1 = Livro("Dom Casmurro", "Machado de Assis", "Romance", 10)
livro2 = Livro("1984", "George Orwell", "Ficção Científica", 5)

lista_livros.append(livro1)
lista_livros.append(livro2)


def cadastrar_livro(titulo, autor, genero, quantidade_disponivel):
    """Cadastra um novo livro na lista de livros."""
    novo_livro = Livro(titulo, autor, genero, quantidade_disponivel)
    lista_livros.append(novo_livro)
    print(f"Livro '{titulo}' cadastrado com sucesso!")

def listar_livros():
    """Exibe todos os livros na lista."""
    if not lista_livros:
        print("A lista de livros está vazia.")
    else:
        print("Lista de Livros:")
        for livro in lista_livros:
            print(f"Título: {livro.titulo}, Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade Disponível: {livro.quantidade_disponivel}")

def buscar_livro(titulo):
    """Busca um livro na lista pelo título."""
    encontrado = False
    for livro in lista_livros:
        if livro.titulo.lower() == titulo.lower():
            print(f"Livro encontrado:")
            print(f"  Título: {livro.titulo}")
            print(f"  Autor: {livro.autor}")
            print(f"  Gênero: {livro.genero}")
            print(f"  Quantidade Disponível: {livro.quantidade_disponivel}")
            encontrado = True
            break  # Stop after finding the first match
    if not encontrado:
        print(f"Livro com o título '{titulo}' não encontrado.")

        !pip install matplotlib

        print("Contagem de livros por gênero:")
for genero, count in contagem_generos.items():
    print(f"{genero}: {count}")

    import matplotlib.pyplot as plt

generos = list(contagem_generos.keys())
quantidades = list(contagem_generos.values())

plt.bar(generos, quantidades)
plt.title("Distribuição de Livros por Gênero")
plt.xlabel("Gênero")
plt.ylabel("Quantidade de Livros")
plt.show()

import matplotlib.pyplot as plt

generos = list(contagem_generos.keys())
quantidades = list(contagem_generos.values())

plt.bar(generos, quantidades)
plt.title("Distribuição de Livros por Gênero")
plt.xlabel("Gênero")
plt.ylabel("Quantidade de Livros")
plt.show()
```
