```python
import sqlite3
import pandas as pd

# Passo 1.2: Conectar ao banco de dados (ou criar um novo)
conexao = sqlite3.connect('dados_vendas.db')
cursor = conexao.cursor()

# Passo 1.3: Criar uma tabela (se não existir)
cursor.execute('''
CREATE TABLE IF NOT EXISTS vendas1 (
id_venda INTEGER PRIMARY KEY AUTOINCREMENT,
data_venda DATE,
produto TEXT,
categoria TEXT,
valor_venda REAL
)
''')

# Passo 1.4: Inserir alguns dados (apenas se a tabela estiver vazia para evitar duplicatas)
cursor.execute("SELECT COUNT(*) FROM vendas1;")
count = cursor.fetchone()[0]
if count == 0:
    cursor.execute('''
    INSERT INTO vendas1 (data_venda, produto, categoria, valor_venda) VALUES
    ('2023-01-01', 'Produto A', 'Eletrônicos', 1500.00),
    ('2023-01-05', 'Produto B', 'Roupas', 350.00),
    ('2023-02-10', 'Produto C', 'Eletrônicos', 1200.00),
    ('2023-03-15', 'Produto D', 'Livros', 200.00),
    ('2023-03-20', 'Produto E', 'Eletrônicos', 800.00),
    ('2023-04-02', 'Produto F', 'Roupas', 400.00),
    ('2023-05-05', 'Produto G', 'Livros', 150.00),
    ('2023-06-10', 'Produto H', 'Eletrônicos', 1000.00),
    ('2023-07-20', 'Produto I', 'Roupas', 600.00),
    ('2023-08-25', 'Produto J', 'Eletrônicos', 700.00),
    ('2023-09-30', 'Produto K', 'Livros', 300.00),
    ('2023-10-05', 'Produto L', 'Roupas', 450.00),
    ('2023-11-15', 'Produto M', 'Eletrônicos', 900.00),
    ('2023-12-20', 'Produto N', 'Livros', 250.00);
    ''')
    # Passo 1.5: Confirmar as mudanças apenas se houver inserções
    conexao.commit()


# Fechar a conexão do cursor para a criação e inserção inicial
cursor.close()


# Reabrir a conexão para a leitura com pandas
conn = sqlite3.connect('dados_vendas.db')
query = "SELECT * FROM vendas1;"
df_vendas = pd.read_sql_query(query, conn)
conn.close()

print(df_vendas.head())
print(df_vendas.info())
print(df_vendas.describe())
print(df_vendas.isnull().sum())

df_vendas.fillna(0, inplace=True)

df_vendas['data_venda'] = pd.to_datetime(df_vendas['data_venda']) #
df_vendas['mes_venda'] = df_vendas['data_venda'].dt.month
total_vendas_por_mes = df_vendas.groupby('mes_venda')['valor_venda'].sum()
vendas_por_categoria = df_vendas.groupby('categoria')['id_venda'].count()
produto_mais_vendido = df_vendas['produto'].value_counts().idxmax()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo dos gráficos
sns.set_style('whitegrid')

# Gráfico de barras para total de vendas por mês
plt.figure(figsize=(10, 6))
sns.barplot(x=total_vendas_por_mes.index, y=total_vendas_por_mes.values, palette='viridis')
plt.title('Total de Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
plt.show()

# Gráfico de barras para vendas por categoria
plt.figure(figsize=(8, 5))
sns.barplot(x=vendas_por_categoria.index, y=vendas_por_categoria.values, palette='magma')
plt.title('Quantidade de Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.show()
```
